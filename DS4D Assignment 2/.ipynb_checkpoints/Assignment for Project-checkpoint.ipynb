# Describe

The date from a social care provider who has developed a tablet and a service called CleverCogs, which provides an easily accessible way for older people to access online games, music and information. 

```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('CC Data (Anonymised safe to share).csv')
print(df1.info())
df1.head()
```

```python
df2 = pd.read_csv('PrimaryConditions age sex by customer reference.csv')
df2.head()
```

There are 2 csv, the first one is more complex and messier compared to the second one. Every row of the first file represents a person living somewhere browses the content of a link at a certain time. And there is more information about this person in the second file, like id, birthdate, gender, and condition. ExternalID is unique, representing an individual, which can be used to connect two sets of data. I mainly analyze data from the perspective of an individual user.

```python
#Clean the data by removing invalid rows and columns and filling invalid values

df1.drop(df1.columns[7:],axis=1,inplace=True)
df1=df1.dropna(subset=['ExternalID'])
df1=df1.drop_duplicates()
df1=df1.dropna(how='all')

df1=df1.fillna(0)

m=df1[df1['Time'].str.contains('<')].index.tolist()
df1=df1.drop(m)

df1.head()
```

```python
#Data characteristics and information

print(df1.dtypes)
print(df1.shape)
print(df1.describe())
print(df1.groupby('LinkTitle').size())

print(df2.dtypes)
print(df2.shape)
print(df2.describe())
```

Both CSV data are tabular. 
The first one has 7 columns and 88443 rows after cleaning, including date, location, text and numbers and there are mixed data types in many columns like there are 3 types of date (float, int, and string) in ExternalID column. The LinkTitle represents the browsed content and many of them are similar (80s Game Shows, 80s Hits, and 80s TV/Gameshows). LinkType shows the type of this LinkTitle, more information of this LinkTitle can be seen in ContentInfo, though most of them are null.
Data of the second CSV is clean and simple with 5 columns and 695 rows including numbers(id or birthdate) and text(condition), which gives the information of disability or impairment of this person.

```python
#Data type conversion and data merge

df1['ExternalID'] = df1['ExternalID'].astype(int)

df=pd.merge(df1, df2)
df=df.fillna(0)

df.to_csv('Entertainment.csv', index=False)
df.tail()
```

```python
#Merged data characteristics

print(df.shape)
print(df.dtypes)
print(df.describe())

print(len(np.unique(df['ExternalID'])))
print(len(df['Building'].drop_duplicates()))
print(len(df['LinkTitle'].drop_duplicates()))
print(len(df['Condition'].drop_duplicates()))

df.head()
```

After the merger, we have a new CSV with 11 columns and 72915 rows, including the information of Time, ExternalID, Role, Building, LinkTitle, LinkType, ContentInfo, CleverCogsUserId, BirthDate, Gender, and Condition in different data types. In total, there are 82 people with 18 conditions living in 18 different placesï¼Œwho have access to 518 different content links.

# Explore

I will look at some general information first, then focus on analyzing users to explore the distribution of these users' features and their activities. I will also select several users to analyze their usage patterns, and figure out whether are some correlations between them.

```python
#Online entertainment distribution over time

import datetime as dt
df['Time']=pd.to_datetime(df['Time'])
df['Weekday']=df['Time'].dt.weekday
Week=df.groupby('Weekday').size()
df['Month']=df['Time'].dt.month
Month=df.groupby('Month').size()

%matplotlib inline
import matplotlib.pyplot as plt     
f,axes = plt.subplots(2,figsize=(6, 8))
Week.plot(color="b", ax=axes[0])
Month.plot(color="r", ax=axes[1])
```

The distribution of online entertainment has a certain pattern in a week. The most frequent use of online entertainment is on Tuesday, then usage frequency keeps falling and reaches the lowest point in the weekend. As for month, the frequency of use is less than 3000 in both November and December, which is the lowest point of the year. And the difference between other months is not obvious while the frequency in February and August is relatively low.

```python
#Distribution of online entertainment link

LinkName=''
for row in df.LinkTitle:
    LinkName+=row+' '
from collections import Counter
freword = Counter(LinkName.split()).most_common(30)
Word=pd.DataFrame(data=freword,columns=['Keyword','Frequency'])
Word.set_index(['Keyword'], inplace=True)
frelink=Counter(df.LinkTitle).most_common(30)
Title=pd.DataFrame(data=frelink,columns=['Linktitle','Frequency'])
Title.set_index(['Linktitle'], inplace=True)

%matplotlib inline
import matplotlib.pyplot as plt     
f,axes = plt.subplots(2,figsize=(15, 9)) 
Word.plot(kind='bar',title='Keyword Distribution of Link',legend=False,ax=axes[0])
Title.plot(kind='bar',title='Title Distribution of Link',legend=False,ax=axes[1])
plt.subplots_adjust(wspace =0, hspace =1)
```

There are more than 500 LinkTitle representing different content, I put them together and calculate the frequency of these worlds, trying to find keywords related to the content. Apparently, Entertainment, Games, and Facebook are top 3, following by Patience, Play, Music, YouTube and BBC, so they prefer games and social networking as entertainment, and they also like to listen to the music, watch videos, and acquire some information related to health and society. Similar information can also be obtained from the second graph.

```python
#Encode non-numerical variables with numerical equivalents

from sklearn import preprocessing
df=df.sort_values(by=['LinkTitle'])
le = preprocessing.LabelEncoder()
le.fit(df['LinkTitle']);
df['LinkId']=le.transform(df['LinkTitle'])

df['Condition'] = df['Condition'].astype(str)
le.fit(df['Condition']);
df['ConditionId']=le.transform(df['Condition'])

df['Building'] = df['Building'].astype(str)
le.fit(df['Building']);
df['BuildingId']=le.transform(df['Building'])

df.tail()
```

```python
#Create a new DataFrame by extracting each user's information

user=df[['ExternalID','BuildingId','Building','BirthDate','Gender','ConditionId','Condition']].drop_duplicates()
print(user.shape)

from collections import Counter
print(Counter(user.ExternalID).most_common(5))
print(user.loc[(user.ExternalID==5967)|(user.ExternalID==5961)|(user.ExternalID==5962)|(user.ExternalID==5964)])
user.drop_duplicates(subset=['ExternalID'],keep='last',inplace=True) 

print(user.shape)
user.head()
```

```python
#Calculate more information and add them to the table

link=df[['ExternalID','LinkTitle']].groupby(by='ExternalID').apply(lambda x:[','.join(x['LinkTitle'])])
link=pd.DataFrame(data=link,columns=['LinkTitle'])
link.reset_index(level='ExternalID')

df['LinkId']=df['LinkId'].astype(str)
link=df[['ExternalID','LinkId']].groupby(by='ExternalID').apply(lambda x:[','.join(x['LinkId'])])
link=pd.DataFrame(data=link,columns=['LinkId'])
link.reset_index(level='ExternalID')

activity=Counter(df['ExternalID']).most_common()
activity=pd.DataFrame(data=activity,columns=['ExternalID','ActiveTimes'])

user=pd.merge(user,link,on='ExternalID')
user=pd.merge(user,activity,on='ExternalID')

user['BirthDate']=pd.to_datetime(user['BirthDate'])
now_year=dt.datetime.today().year
user['Age']=now_year-user.BirthDate.dt.year

print(user.describe())
user.head()
```

```python
#Use K-Means to do clustering

c=user[['ConditionId','ActiveTimes']]
c= np.array(c)
c=c.tolist()

from sklearn.cluster import KMeans
from sklearn.externals import joblib
from sklearn import cluster

n=KMeans(n_clusters=3)
res=n.fit_predict(c)
lablel=n.labels_
center=n.cluster_centers_
inertia=n.inertia_

print(lablel)
print(center)
print(inertia)

for i in range(len(c)):
    if int(lablel[i])==0:
        plt.scatter(c[i][0],c[i][1],color='red')
    if int(lablel[i])==1:
        plt.scatter(c[i][0],c[i][1],color='green')
    if int(lablel[i])==2:
        plt.scatter(c[i][0],c[i][1],color='blue')
```

```python
#Plot the correlations between different variables

import seaborn as sns          
%matplotlib inline
from matplotlib import pyplot as plt
sns.pairplot(user,hue='Gender',vars=['ActiveTimes','Age','BuildingId','ConditionId'])

user.describe()
user.to_csv('user.csv', index=False)
```

In order to analyze from the perspective of users, I created a new DataFrame with each user's information including calculated age and number of active times. I also extracted links that everyone has viewed to further analyze entertainment preferences and patterns. After plotting, we could find an outlier in the age which is 1, and users between the ages of 60 and 70 especially females use online entertainment more frequently than other age groups, Elderly Care or Cerebral Palsy are more likely to use this service. There aren't too many differences between male and female in the distribution of Building, Condition or Age.

```python
#Select users with similar information to compare their online entertainment preference

sample=user.loc[(user.Condition=='Cerebral Palsy')&(user.Gender=='M')&(user.ActiveTimes>500)&(user.Age>50)]
print(sample)

A=user.loc[(user.ExternalID==5541)]
Alist=(A.LinkId).astype(str)
A=np.array(Alist).tolist()
Astr=",".join(A)
B=user.loc[(user.ExternalID==4763)]
Blist=(B.LinkId).astype(str)
B=np.array(Blist).tolist()
Bstr=",".join(B)

import difflib
seq = difflib.SequenceMatcher(None, Astr, Bstr)
seq.ratio()
```
